When you suspect the relationship between features and the target variable might be non-linear and potentially vary locally across different data regions.

#### Advantageous situation:
When you have a mix of linear and non-linear relationships between features and the target variable, and want more flexibility in modeling these relationships.

## Question 3 

Define the following types of smooth functions from PYGAM library (https://pygam.readthedocs.io/en/latest/notebooks/tour_of_pygam.html ) and list at least one relevant potential use: 

a) l() linear terms

b) s() spline terms

c) f() factor terms

d) te() Tensor product interaction terms


### a) l() linear terms

#### Definition: 
The l() function represents linear terms in a Generalized Additive Model (GAM). It models a linear relationship between a continuous feature (numerical variable) and the target variable.
#### Use Case:
When you have features with a proportional relationship to the target variable. For example, modeling house prices as a function of square footage (assuming price increases linearly with increasing square footage).

### b) s() spline terms
#### Definition: 

The s() function represents smooth spline terms in a GAM. It allows for modeling non-linear relationships between a feature and the target variable. Splines are piecewise polynomial functions that provide flexibility in capturing smooth curves.
#### Use Case:
When you suspect a non-linear trend in the data, like modeling sales figures over time (which may exhibit seasonal patterns) or customer churn probability based on a customer's age (which might not be a simple linear increase).
### c) f() factor terms

#### Definition: The f() function represents categorical features (also known as factor variables) in a GAM. It encodes the categories into a set of dummy variables, allowing the model to capture different effects for each category on the target variable.
#### Use Case:
When you have categorical predictors that might influence the target variable in different ways. For example, modeling customer satisfaction based on customer type (bronze, silver, gold) where each category might have a distinct satisfaction level.
### d) te() Tensor product interaction terms

#### Definition: The te() function represents tensor product interaction terms in a GAM. It allows modeling the interaction effect between two or more features. The interaction term captures how the effect of one feature on the target variable changes depending on the level of another feature.
#### Use Case:
When you suspect that the effect of one feature on the target variable depends on the level of another feature. For example, modeling website click-through rate (CTR) considering both user demographics (age) and product category. The CTR for a particular product category might vary depending on the user's age group.
By combining these smooth functions in a GAM, you can build flexible and powerful models that capture complex relationships between features and the target variable.

## Question 4

Briefly explain what spatial and temporal autocorrelation are? List one way in which GAMs could help to account for this (e.g. indicate the type of smooth function you would use or potential types of correlation structures).



### Spatial Autocorrelation:

Spatial autocorrelation describes the clustering or patterning of a variable across geographic locations. It measures the degree to which nearby observations are more similar (positive autocorrelation) or dissimilar (negative autocorrelation) compared to random chance.

Example: Disease outbreaks in a city might exhibit positive spatial autocorrelation, indicating higher infection rates in neighboring areas.

### Temporal Autocorrelation:

Temporal autocorrelation describes the correlation of a variable over time. It measures the degree to which observations at close time points are more similar (positive autocorrelation) or dissimilar (negative autocorrelation) compared to random chance.

Example: Stock prices often exhibit positive temporal autocorrelation, meaning past price movements can influence future movements.

### Using GAMs to Account for Autocorrelation

GAMs (Generalized Additive Models) can be a powerful tool to account for autocorrelation issues in both spatial and temporal data analysis. Here's how:

#### 1. Smoothing with Spline Terms (s())

Spline terms in GAMs allow for modeling non-linear trends across space or time. This flexibility can capture underlying spatial or temporal patterns that might contribute to autocorrelation.
For example, a spline term for location coordinates in a disease outbreak model could capture spatial clustering, while a spline term for time in a stock price model could capture temporal trends and seasonality.

#### 2. Interaction Terms (te())

Tensor product interaction terms (te()) can incorporate spatial or temporal coordinates as interacting features. This allows the model to capture how the effect of other variables on the target variable changes depending on location or time.
For example, you could create an interaction term between a spatial coordinate and a pollution source variable to model how pollution levels vary depending on location relative to the source. Similarly, you could interact a time variable with a weather variable to account for how weather seasonality affects sales.
